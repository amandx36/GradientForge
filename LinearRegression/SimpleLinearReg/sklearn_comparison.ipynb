```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
doc = pd.read_csv("./placement.csv")
# !ls
```

```python
doc.head()
```

# Plotting the graph 

```python
plt.scatter(doc["cgpa"], doc["package"], color="green")
plt.xlabel("CGPA")
plt.ylabel("Package")
plt.title("CGPA vs Package")
plt.show()
```

```python
doc.head(20)
```

```python
x = doc.iloc[:,0:1]
print(x)
```

```python
# only the date in second column 
y = doc.iloc[:,1:2]
print(y)
```

```python

from sklearn.model_selection import train_test_split
x_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)

```

```python

from sklearn.linear_model import LinearRegression 
```

```python

lr = LinearRegression()
```

```python
# training 
lr.fit(x_train , y_train)
```

```python

X_test
```

```python

y_test.head()
```

```python

lr.predict(X_test)
```

```python
# graph 
plt.scatter(doc["cgpa"], doc["package"], color="red")
#  x tain vs  predicted y train
plt.plot(x_train , lr.predict(x_train),color= "blue" )
```

```python

m = lr.coef_
print(m)
```
